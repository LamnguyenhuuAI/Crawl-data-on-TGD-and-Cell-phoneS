## Phân tích giá laptop ở CellphoneS

```python
cell_laptop['Price'].min()
```

```python
cell_laptop['Price'].max()
```

```python
a = cell_laptop[cell_laptop['Price'] <= 10000000.0] #Phân khúc giá thấp dưới 10tr VND
tb = len(a)/len(cell_laptop)*100
tb
```

```python
b = cell_laptop[(cell_laptop['Price'] > 10000000.0) & (cell_laptop['Price'] <= 30000000.0)] # Phân khúc giá laptop trung bình và cận cao cấp 
can_cao = len(b)/len(cell_laptop)*100
can_cao
```

```python
c = cell_laptop[cell_laptop['Price'] > 30000000.0] #Phân khúc laptop cao cấp
cao = len(c)/len(cell_laptop)*100
cao
```

```python
y = np.array([tb,can_cao,cao])
mylabels = ['Thấp', "Trung Bình và Cận Cao Cấp", "Cao Cấp"]
exp = [0.1,0.1,0.1]
font1 = {'family':'serif','color':'red','size':13}
plt.pie(y, labels = mylabels,explode = exp,startangle = 90,shadow = True,autopct='%1.1f%%')
plt.title('Biểu đồ số lượng laptop theo mức giá ở CellphoneS',loc = 'center',fontdict = font1)
plt.show() 
```

## Phân tích giá laptop ở Thế giới di động

```python
a = tgdd_laptop[tgdd_laptop['Price'] <= 10000000.0]
tb = len(a)/len(tgdd_laptop)*100
```

```python
b = tgdd_laptop[(tgdd_laptop['Price'] > 10000000.0) & (tgdd_laptop['Price'] <= 30000000.0)] # Phân khúc giá laptop trung bình và cận cao cấp 
can_cao = len(b)/len(tgdd_laptop)*100
```

```python
c = tgdd_laptop[tgdd_laptop['Price'] > 30000000.0] #Phân khúc laptop cao cấp
cao = len(c)/len(tgdd_laptop)*100
```

```python
y = np.array([tb,can_cao,cao])
mylabels = ['Thấp', "Trung Bình và Cận Cao Cấp", "Cao Cấp"]
exp = [0.1,0.1,0.1]
font1 = {'family':'serif','color':'blue','size':13}
plt.pie(y, labels = mylabels,explode = exp,startangle = 90,shadow = True, autopct='%1.1f%%')
plt.title('Biểu đồ số lượng laptop theo mức giá ở TGDĐ',loc = 'center',fontdict = font1)
plt.show() 
```

## Phân tích laptop ở thế giới di động 

```python
a = tgdd_laptop['Ram'].value_counts()
fig = plt.figure(figsize = (10, 5))
plt.bar(a.keys(), a.values,
        width = 0.4)
plt.xlabel("Các loại Ram")
plt.ylabel("Count")
plt.title("Biểu đồ Ram laptop ở thế giới di động")
plt.show()
```

```python
b = tgdd_laptop['Rom'].value_counts()
fig = plt.figure(figsize = (10, 5))
plt.bar(b.keys(), b.values, color ='maroon',
        width = 0.4)
plt.xlabel("Các loại Rom")
plt.ylabel("Count")
plt.title("Biểu đồ Rom laptop ở thế giới di động")
plt.show()
```

```python
a = tgdd_laptop['Cpu'].value_counts()
fig = plt.figure(figsize = (10, 5))
plt.barh(a.keys(), a.values, color ='maroon',)
plt.xlabel("Count")
plt.ylabel("Các loại Chip")
plt.title("Biểu đồ Chip của laptop ở thế giới di động")
plt.show()
```
