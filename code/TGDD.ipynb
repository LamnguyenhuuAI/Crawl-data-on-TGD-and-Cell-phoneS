```python
dt = pd.read_csv('./DienThoaiTGDD.csv')
dt
```

```python
a = dt['Links']
```

```python
name_ = []
price_ = []
os_ = []

for i in a:
    driver = webdriver.Edge(EdgeChromiumDriverManager().install())
    driver.get('%s' % i)  # Lấy link sản phẩm
    time.sleep(2)  # Thời gian chờ
    driver.implicitly_wait(10)  # Thời gian chờ xử lý bất đồng bộ
    
    try:
        name = driver.find_element(By.XPATH, '/html/body/section[1]/h1').text
        price = driver.find_element(By.CLASS_NAME, 'box-price-present').text  # Lấy giá của sản phẩm
        
        parameter_list = driver.find_element(By.CLASS_NAME, 'parameter__list')
        parameter_items = parameter_list.find_elements(By.TAG_NAME, 'li')
        
        os = ''
        for item in parameter_items:
            p_element = item.find_element(By.TAG_NAME, 'p')
            if p_element.text == 'Hệ điều hành:':
                os_element = item.find_element(By.TAG_NAME, 'span')
                os = os_element.text
                break
        
    except:  # Nếu không tìm thấy element
        try:
            name = driver.find_element(By.CLASS_NAME, 'tab').text
            price = 'N/A'  # Tìm tiếp giá của sản phẩm
            
            parameter_list = driver.find_element(By.CLASS_NAME, 'parameter__list')
            parameter_items = parameter_list.find_elements(By.TAG_NAME, 'li')
            
            os = ''
            for item in parameter_items:
                p_element = item.find_element(By.TAG_NAME, 'p')
                if p_element.text == 'Hệ điều hành:':
                    os_element = item.find_element(By.TAG_NAME, 'span')
                    os = os_element.text
                    break
            
        except:
            name = 'N/A'
            os = 'N/A'

    name_.append(name)
    price_.append(price)
    os_.append(os)

    driver.implicitly_wait(10)  # Thời gian chờ xử lý bất đồng bộ
    driver.close()
```

```python
df = pd.DataFrame(list(zip(name_,price_, os_,)), columns= ['Name', 'Price', 'OS'])
df
```
