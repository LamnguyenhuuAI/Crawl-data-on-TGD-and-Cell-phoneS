```python
from selenium import webdriver
import pandas as pd
import numpy as np
import time
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from webdriver_manager.chrome import ChromeDriverManager
```

# Lấy link điện thoại

```python
driver = webdriver.Chrome(ChromeDriverManager().install())
url = 'https://www.thegioididong.com/dtdd'
driver.get(url)
```

```python
elems = driver.find_elements(By.CSS_SELECTOR,".listproduct [href]")
links = [elem.get_attribute('href') for elem in elems]
print(links)
```

## Lấy ra những link không phải https://www.thegioididong.com/dtdd# và javascript:void(0)

```python
links = [s for s in links if s != 'https://www.thegioididong.com/dtdd#']
links = [s for s in links if s != 'javascript:void(0)']
print(links) # Loại bỏ xâu không phải là link sản phẩm
```

```python
df1 = pd.DataFrame(links, columns= ['Links']) #Đưa link vào bảng dữ liệu 
df1
```

```python
df1.to_csv('LinkDienThoai(TGDĐ).csv') #lưu link sản phẩm
```

```python
name_ = []
price_ = []
for i in links: # Duyệt tất cả các link có trong danh sách
    driver = webdriver.Chrome(ChromeDriverManager().install())
    driver.get('%s'%i) #Lấy link sản phẩm
    time.sleep(2) #Thời gian chờ
    driver.implicitly_wait(10) # Thời chờ xử lí bất đồng bộ
    try:
        name = driver.find_element(By.XPATH,'/html/body/section[1]/h1').text # Lấy tên sản phẩm
        price = driver.find_element(By.CLASS_NAME,'box-price-present').text # Lấy giá của sản phẩm
    except: #Nếu không tìm thấy element
        try:
            name = 'N/A' #Gán name bằng rỗng
            price = driver.find_element(By.XPATH,'//*[@id="tab235838"]/div/div[1]/b/i').text #Tìm tiếp giá của sản phẩm
        except:
            price = 'N/A' #Gán giá là rỗng
    name_.append(name)
    price_.append(price)
    driver.implicitly_wait(10) # Thời chờ xử lí bất đồng bộ
    driver.close()
```

```python
price_
```

```python
data = pd.DataFrame({'Name':name_,
                    'Price': price_})
data
```

```python
data.to_csv('DienThoai.csv')
```
